## http://openapi.seoul.go.kr:8088/(인증키)/xml/CardBusStatisticsServiceNew/1/5/20151101/6617/

```python
TYPE = 'json'
SERVICE = 'CardBusStatisticsServiceNew'
START_INDEX = str(1)#데이터 행 시작
END_INDEX = str(100) # 데이터 행 끝
USE_DT = '20231001' #YYYYMMDD format
BUS_ROUTE_NO = '7016'
```

```python
params = "/".join([KEY,TYPE,SERVICE,START_INDEX,END_INDEX,USE_DT,BUS_ROUTE_NO])
params
```

```python
_URL = 'http://openapi.seoul.go.kr:8088'
URL = "/".join([_URL,params])
print(URL)
```

```python
import requests as req

busStation = req.get(URL).json()
busStation[SERVICE]['list_total_count']
```

## USE_DT       
## RIDE_PASGR_NUM
##  ALIGHT_PASGR_NUM   
## BUS_STA_NM



```python
sum_RIDE_PASGR_NUM = 0

for bus in busStation[SERVICE]['row']:

  print('{0}\t {1}\t {2}\t {3}\t'.format(bus['USE_DT'],bus['RIDE_PASGR_NUM'],bus['ALIGHT_PASGR_NUM'],bus['BUS_STA_NM']))
  sum_RIDE_PASGR_NUM += bus['RIDE_PASGR_NUM']

print('승차인원 합계: {0}'.format(sum_RIDE_PASGR_NUM))
```
