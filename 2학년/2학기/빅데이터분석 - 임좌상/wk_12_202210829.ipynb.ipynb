```python
import numpy as np

x=np.random.normal(15, 3, size=200)
```

```python
x[:10]
```

```python
from scipy import stats

xZscore=stats.zscore(x)
```

```python
xZscore
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].hist(x, bins=20)
ax[0].set_title('xGraph')
ax[0].set_xlabel('x')
ax[0].set_ylabel('Frequency')

ax[1].hist(xZscore, bins=20)
ax[1].set_title('xZscoreGraph')
ax[1].set_xlabel('Zscore')
ax[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
