```python
total_df['총생활인구수_M2'] = total_df['총생활인구수'].shift(-1416)
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(total_df, test_size=1416, shuffle=False)
```

```python
train_set.dropna(axis=0, inplace=True)
train_x = train_set.drop('총생활인구수_M2', axis=1)
train_y = train_set['총생활인구수_M2']

test_x = test_set.drop('총생활인구수_M2', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import *
```

```python
scaler = MinMaxScaler()
train_x_s = scaler.fit_transform(train_x)
test_x_s = scaler.transform(test_x)
```

```python
from catboost import CatBoostRegressor, Pool
```

```python
cat = CatBoostRegressor()
```

```python
cat.fit(train_x_s, train_y.ravel())
cat_pred = cat.predict(test_x_s)
```

```python
cat_pred = pd.DataFrame({'count': cat_pred})
cat_pred.to_csv('dkdkdsfksljlkThrTry_Catboost1.csv', index_label='id')
```
