```python
#아래에 실습코드를 작성하고 결과를 확인합니다.
data['1h'] = data['총생활인구수'].shift(1)
data['1d'] = data['총생활인구수'].shift(24)
data['7d'] = data['총생활인구수'].shift(24*7)
data['14d'] = data['총생활인구수'].shift(24*14)
data['30d'] = data['총생활인구수'].shift(24*30)
data['90d'] = data['총생활인구수'].shift(24*90)
data['180d'] = data['총생활인구수'].shift(24*180)
data['1y'] = data['총생활인구수'].shift(24*365)
data['2y'] = data['총생활인구수'].shift(24*730)
data['3y'] = data['총생활인구수'].shift(24*1095)
```

```python
data
```

```python
plt.figure(figsize=(40,10))
plt.plot(data)
```

```python
data = data.dropna()
```

```python
data
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
x = data.iloc[ :,1:]
y = data.iloc[ :,:1]
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
x = x.reset_index(drop=True)
y = y.reset_index(drop=True)
```

```python
train_x = x[:len(y)]
```

```python
train_y = y[:len(y)]
```

```python
train_x
```

```python
train_y
```

```python
#2021년10월 ~ 2021년11월
test_x = x[len(y)-((59*24)+31*24):len(y)-744]
```

```python
test_x
```

```python
#2021년10월 ~ 2021년11월
test_y = y[len(y)-((59*24)+31*24):len(y)-744]
```

```python
test_x = test_x.reset_index(drop=True)
test_y = test_y.reset_index(drop=True)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
```

```python
print('훈련 데이터의 크기 :', train_x.shape, train_y.shape)
print('테스트 데이터의 크기 :', test_x.shape, test_y.shape)
```

```python
train_x = train_x.values
train_y = train_y.values
```

```python
# 아래에 실습코드를 작성하세요.

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score as r2
from math import sqrt

LR_reg = LinearRegression()
LR_reg.fit(train_x, train_y)
```

```python
# Test 데이터로 성능을 평가해보세요.

y_pred_LR = LR_reg.predict(test_x)

print('LinearRegression')
print("Mean Squared Error : {0:.5f}".format(mse(test_y,y_pred_LR)))
print("Root Mean Squared Error : {0:.5f}".format(sqrt(mse(test_y,y_pred_LR))))
print("r2 : ", r2(test_y,y_pred_LR))
```

```python
result = pd.DataFrame(y_pred_LR)
result
```
