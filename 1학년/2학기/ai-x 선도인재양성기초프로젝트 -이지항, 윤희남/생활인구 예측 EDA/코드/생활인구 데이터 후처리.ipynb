```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

df19 = pd.read_csv('test2019.csv', sep=',', index_col=0)
df20 = pd.read_csv('test2020.csv', sep=',', index_col=0)
df21 = pd.read_csv('test2021.csv', sep=',', index_col=0)
```

```python
df_list = [df19, df20, df21]
total_df = pd.concat(df_list)
total_df
```

```python
total_df['기준일ID'] = pd.to_datetime(total_df['기준일ID'].astype(str), format='%Y%m%d')
total_df
```

```python
plt.figure(figsize=(50, 20))
plt.rc('xtick', labelsize=50)  # x축 눈금 폰트 크기 
plt.rc('ytick', labelsize=50)  # y축 눈금 폰트 크기
plt.plot(total_df['기준일ID'], total_df['총생활인구수'])
plt.show()
```

```python
total_df['year'] = total_df['기준일ID'].dt.year
total_df['month'] = total_df['기준일ID'].dt.month
total_df['day'] = total_df['기준일ID'].dt.day
```

```python
total_df.rename(columns = {'시간대구분' : 'hour'}, inplace = True)
```

```python
total_df = total_df.loc[:, ['기준일ID','year', 'month', 'day', 'hour', '총생활인구수']].copy()
```

```python
total_df['1d'] = total_df['총생활인구수'].shift(24)
total_df['7d'] = total_df['총생활인구수'].shift(24*7)
total_df['2m'] = total_df['총생활인구수'].shift(1416)
total_df['2m1d'] = total_df['총생활인구수'].shift(1416+24)
total_df['2m7d'] = total_df['총생활인구수'].shift(1416+24*7)

total_df['총생활인구수_MA_1W'] = total_df['총생활인구수'].rolling(168).mean() 
total_df['총생활인구수_MA_2W'] = total_df['총생활인구수'].rolling(336).mean() 
total_df['총생활인구수_MA_1M'] = total_df['총생활인구수'].rolling(720).mean() 
total_df['총생활인구수_MA_2M'] = total_df['총생활인구수'].rolling(1440).mean() 
```
