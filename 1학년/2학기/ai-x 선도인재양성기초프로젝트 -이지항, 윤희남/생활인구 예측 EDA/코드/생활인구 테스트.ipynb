```python
df17 = pd.read_csv('test2017.csv', sep=',', index_col=0)
df18 = pd.read_csv('test2018.csv', sep=',', index_col=0)
df19 = pd.read_csv('test2019.csv', sep=',', index_col=0)
df20 = pd.read_csv('test2020.csv', sep=',', index_col=0)
df21 = pd.read_csv('test2021.csv', sep=',', index_col=0)
```

```python
df_total = pd.concat([df17,df18,df19,df20,df21])
```

```python
df_total.shape
```

```python
df_total.info()
```

```python
df_total['기준일ID'] = pd.to_datetime(df_total['기준일ID'], format='%Y%m%d')

df_total['year']=df_total['기준일ID'].dt.year
df_total['month']=df_total['기준일ID'].dt.month
df_total['day']=df_total['기준일ID'].dt.day
```

```python
df_total.rename(columns={'시간대구분': 'hour'}, inplace=True)
df_total = df_total.loc[:, ['year', 'month', 'day', 'hour', '총생활인구수']].copy()
```

```python
df_total['총생활인구수_M2'] = df_total['총생활인구수'].shift(-1416)
```

```python
df_total['총생활인구수_MA_1W'] = df_total['총생활인구수'].rolling(168).mean() 
df_total['총생활인구수_MA_2W'] = df_total['총생활인구수'].rolling(336).mean() 
df_total['총생활인구수_MA_1M'] = df_total['총생활인구수'].rolling(720).mean() 
df_total['총생활인구수_MA_2M'] = df_total['총생활인구수'].rolling(1440).mean() 
```

```python
df_total['1d'] = df_total['총생활인구수'].shift(24)
df_total['7d'] = df_total['총생활인구수'].shift(24*7)
```

```python
cols = ['year', 'month', 'day', 'hour']

df_total.drop(cols, axis=1, inplace=True)
df_total.reset_index(drop=True, inplace=True)
```

```python
df_total.head()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df_total, test_size=1416, shuffle=False)
```

```python
import pandas as pd
import numpy as np
# 학습/평가 데이터 로딩
train_x = pd.read_csv("train_x_js.csv")
train_y = pd.read_csv("train_y_js.csv")
test_x = pd.read_csv("test_x_js.csv")
```

```python
train_x.shape
train_y.shape
```

```python
train_set.dropna(axis=0, inplace=True)
train_x = train_set.drop('총생활인구수_M2', axis=1)
train_y = train_set['총생활인구수_M2']

test_x = test_set.drop('총생활인구수_M2', axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import *
```

```python
scaler = MinMaxScaler()
train_x_s = scaler.fit_transform(train_x)
test_x_s = scaler.transform(test_x)
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(train_x_s, train_y.ravel())
```

```python
pred_rf = rf.predict(test_x_s)

pred_rf = pd.DataFrame({'count': pred_rf})
pred_rf.to_csv('FirstTry_RandomForest.csv', index_label='id')
```
