# 데이터 전처리

```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
sns.set(font="Malgun Gothic",#"NanumGothicCoding", 
        rc={"axes.unicode_minus":False}, # 마이너스 부호 깨짐 현상 해결
        style='darkgrid')
```

```python
df_total = pd.read_csv('df_total.csv')
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.

df_total['기준일ID'] = pd.to_datetime(df_total['기준일ID'])

df_total['year'] = df_total['기준일ID'].dt.year
df_total['month'] = df_total['기준일ID'].dt.month
df_total['day'] = df_total['기준일ID'].dt.day
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.

df_total = df_total[['year','month','day','시간대구분','총생활인구수']]
df_total.rename(columns={'시간대구분':'hour'},inplace=True)
```

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
df_total = df_total.drop(['year','month','day','hour'], axis=1)
```

```python
#이상치 제거
pct25 = 37917.393025
pct75 = 43514.610950
iqr = pct75 - pct25
Pct25 = pct25 - 1.5*iqr
Pct75 = pct75 + 1.5*iqr
```

```python
#하한
Pct25
```

```python
#상한
Pct75
```

```python
df_total['총생활인구수'] = np.where((df_total['총생활인구수'] < Pct25) | (df_total['총생활인구수'] > Pct75),
                                  np.nan, df_total['총생활인구수'])

df_total['총생활인구수'].isna().sum()
```

```python
#결측값 전날값으로 채우기
df_total = df_total.fillna(method='ffill') 
```

```python
plt.figure(figsize=(30,10))
plt.plot(df_total, color='r')
```

```python
df_total.to_csv('data_processing.csv', index=False)
```
